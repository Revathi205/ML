```python

import numpy as np
import pandas as pd

from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# -------------------------------
# Load built-in dataset
# -------------------------------
data = load_breast_cancer()

X = pd.DataFrame(data.data, columns=data.feature_names)
y = data.target

print("Dataset shape:", X.shape)
# -------------------------------
# Train-test split
# -------------------------------
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.1 , random_state=42
)

# -------------------------------
# AdaBoost Model
# -------------------------------
model = AdaBoostClassifier(
    estimator=DecisionTreeClassifier(max_depth=1),
    n_estimators=50,
    learning_rate=1.0,
    random_state=42
)

# Train
model.fit(X_train, y_train)

# Predict
y_pred = model.predict(X_test)

# -------------------------------
# Evaluation
# -------------------------------
print("\nAccuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))
```

### Advantages of Boosting (e.g., Gradient Boosting, AdaBoost)

Boosting algorithms like Gradient Boosting and AdaBoost are powerful ensemble methods that sequentially build models, where each new model attempts to correct the errors of the previous ones. Here are their key advantages:

1.  **High Accuracy**: Boosting often achieves very high accuracy and is considered one of the best out-of-the-box predictors for many types of data. It can capture complex non-linear relationships in the data.
2.  **Error Correction**: By focusing on the misclassified instances of previous models, boosting iteratively reduces bias, leading to more robust models.
3.  **Feature Importance**: Boosting algorithms can provide insights into feature importance, indicating which features contribute most to the predictions.
4.  **Handles Various Data Types**: They can work effectively with numerical and categorical features.

### Comparison with Bootstrap (Bagging) Algorithms (e.g., Random Forest)

**Bagging** (Bootstrap Aggregating) builds multiple independent models (e.g., decision trees) on different bootstrap samples (random samples with replacement) of the training data. The final prediction is an average (for regression) or majority vote (for classification) of these models. Random Forest is a prominent example of bagging.

*   **Key Difference**: Bagging aims to reduce variance by averaging independent models. Boosting aims to reduce bias and convert weak learners into strong learners by sequentially correcting errors.
*   **Parallelization**: Bagging algorithms can be easily parallelized since models are built independently. Boosting is inherently sequential.
*   **Overfitting**: Bagging, especially Random Forests, is less prone to overfitting than boosting, though boosting can also be regularized.

### Comparison with Simple Decision Tree Algorithms

A **Simple Decision Tree** is a single tree-based model that makes decisions by recursively splitting the data based on feature values. They are intuitive and easy to interpret but can be prone to overfitting and may not capture complex patterns well.

*   **Accuracy**: Ensemble methods (boosting and bagging) almost always outperform a single decision tree, especially on complex datasets.
*   **Overfitting**: Simple decision trees are highly susceptible to overfitting, especially deep ones. Ensemble methods mitigate this by combining multiple models.
*   **Bias-Variance Trade-off**: A single decision tree can have high variance. Bagging reduces variance, while boosting reduces bias and can also reduce variance when properly regularized.
